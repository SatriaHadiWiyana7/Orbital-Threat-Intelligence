```python
import pandas as pd
import sys
import matplotlib.pyplot as plt
import seaborn as sns

# Menambahkan path scripts agar bisa diimport
sys.path.append('../scripts')
from eda_helper import clean_column_names, plot_missing_values

# Menggunakan low_memory=False karena dataset >300MB dengan mixed types
path = r'../data/raw/dataset.csv'
df = pd.read_csv(path, low_memory=False)
df = clean_column_names(df)

print(f"Dataset berhasil dimuat: {df.shape[0]} baris & {df.shape[1]} kolom")
```

```python
# Cek Missing Values menggunakan helper
missing_report = plot_missing_values(df)
print(missing_report)
```

```python
plt.figure(figsize=(8, 5))
# Menghapus NaN pada kolom pha untuk visualisasi target yang bersih
sns.countplot(x='pha', data=df.dropna(subset=['pha']), palette='viridis')
plt.title('Distribusi Kelas Potentially Hazardous Asteroid (PHA)')
plt.show()

# Menampilkan angka pasti
print(df['pha'].value_counts())
```

```python
# Memilih fitur orbital utama yang datanya lengkap 100%
orbital_features = ['h', 'e', 'a', 'q', 'i', 'om', 'w', 'ma', 'ad', 'n', 'tp', 'per_y', 'moid_ld']
plt.figure(figsize=(12, 10))
sns.heatmap(df[orbital_features].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap Korelasi Parameter Orbital')
plt.show()
```
